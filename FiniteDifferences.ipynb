Basis functions (hat functions)

```julia
function hat(i,N,x::T) where {T}
    @assert 0<= i <N
    #calculate grid spacing
    h=1/T(N-1)
    #collocation points
    xim1=h*(i-1)
    xi=h*i
    xip1=h*(i+1)
    #case distinction (piecewise linear!)
    if x <= xim1
        return T(0)
    elseif x<= xi
        return (x-xim1)/h #slope
    elseif x<=xip1
        return -(x-xip1)/h
    else
        return T(0)
    end
end
```

```julia
plot(x=xs,y=[hat(5,11,x) for x in xs],Geom.line)
```

## Discrete functions

```julia
struct Fun{T}
    coeffs::Vector{T}
end
```

```julia
dump(Fun{Float64}([1,2,3]))
```

```julia
function Base.:+(f::Fun{T},g::Fun{T}) where {T}  #doesn't need to do this if just use vector type and don't define your own type T
    Fun{T}(f.coeffs+g.coeffs)
end
```

```julia
function Base.:-(f::Fun{T},g::Fun{T}) where {T}
    Fun{T}(f.coeffs-g.coeffs)
end
```

```julia
function sample(::Type{T},f,N) where {T} #turns smooth function into discrete one, N is number of collocation points
    h=1/T(N-1)
    coeffs=T[f(h*i) for i in 0:N-1]
    Fun{T}(coeffs)
end    
```

```julia
sample(Float64,x->x^2,5)
```

Specifies type explicitly

```julia
function sample(f,N) #turns smooth function into discrete one, N is number of collocation points
    h=1/(N-1)
    coeffs=Float64[f(h*i) for i in 0:N-1]
    Fun{Float64}(coeffs)
end    
```

```julia
sample(x->x^2,5)
```

```julia
g(x)=exp(-40*(x-0.6)^2)
```

```julia
g11coeffs=sample(g,11)
```

```julia
function evaluate(f::Fun{T},x) where {T}
    N=length(f.coeffs)
    sum(f.coeffs[i+1]*hat(i,N,x) for i in 0:N-1)
end
```

```julia
g11(x)=evaluate(g11coeffs,x)
```

```julia
plot(x=xs,y=[g11(x) for x in xs],Geom.line)
```

```julia
g21coeffs=sample(g,21)
g21(x)=evaluate(g21coeffs,x)
plot(x=xs,y=[g21(x) for x in xs],Geom.line)
```

## Quadrature

```julia
function quad(f::Fun{T}) where {T}
    N=length(f.coeffs)
    h=1/T(N-1)
    #intialize sum to zero
    s=T(0)
    for i in 0:N-1
        #boundary basis functions have half the area
        if i==0 || i==N-1
            w=T(0.5)
        else
            w=T(1)
        end
        #w=(i==0 || i ==N-1 ? T(0.5) : T(1) *h)
        #add up contributions from all basis functions
        s += w*h*f.coeffs[i+1]
    end
    s
end
```

```julia
quad(g11coeffs)
```

```julia
quad(g21coeffs)
```

## L2 norm

```julia
function l2norm(f::Fun{T}) where {T}
    coeffs_abs2=[abs(c)^2 for c in f.coeffs]
    sqrt(quad(Fun{T}(coeffs_abs2)))
end
```

```julia
l2norm(g11coeffs)
```

```julia
l2norm(g21coeffs)
```
