## Types

```julia
struct Point
    x::Float64
    y::Float64
end
```

```julia
Point(1,2)
```

```julia
function pointadd(p::Point, q::Point)
    Point(p.x + q.x,p.y+q.y)
end
```

```julia
pointadd(Point(1,2),Point(3,4))
```

```julia
function Base.:+(p::Point,q::Point)
    Point(p.x + q.x, p.y + q.y)
end
```

## Generic programming

```julia
function add2(x::T,y::T) where {T<:Number} #require x, y to be of type T i.e. a number
    x + y
end
```

```julia
add2(3,4)
```

```julia
add2(3.0,4.0)
```

```julia
add2(3,4.0) #error because not same type
```

```julia
@code_native add2(1,2)
```

```julia
@code_native add2(1.,2.)
```

```julia
@code_native add2(1.0im,2.0im)
```
